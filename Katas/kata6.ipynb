# Ejercicio 2: Trabajando con datos de una lista
## Usar slices para recuperar partes de una lista

```python
# Lista de planetas
planets = ["Mercurio", "Venus", "Tierra", "Marte", "Júpiter", "Saturno", "Urano", "Neptuno"]
```

```python
# Solicitamos el nombre de un planeta *Pista:  input()*
entrada = input("Ingresa el nombre de un planeta, procura que empiece con MAYÚSCULA")
print(entrada)
```

```python
# Busca el planeta en la lista
lugar_planeta=planets.index(entrada)
print(f"El planeta está en el lugar {lugar_planeta} de la lista planetas")
```

```python
# Muestra los planetas más cercanos al sol
pl_cerc = planets[0:lugar_planeta]
print(f"Los {len(pl_cerc)} planetas más cercanos al Sol que el planeta {entrada} son {pl_cerc}.")
```

```python
# Muestra los planetas más lejanos al sol
pl_lej = planets[lugar_planeta+1:]
print(f"Los {len(pl_lej)} planetas más lejanos al Sol que el planeta {entrada} son {pl_lej}.")
```
