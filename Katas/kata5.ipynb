
# Ejercicio1 - Utilizar operadores aritméticos
## Operadores aritméticos en Python

```python
# Crear variables para almacenar las dos distancias
# ¡Asegúrate de quitar las comas!

dTie = 149597870 #Distancia de la Tierra al Sol en km
dJup = 778547200 #Distancia de Júpiter al Sol en km

conv = 0.621

print(f"La distancia del Sol a la Tierra es de {dTie*conv} millas.")
print(f"La distancia del Sol a Júpiter es de {dJup*conv} millas.")

# Calcular la distancia entre planetas
dJT = dJup - dTie
print(f"La distancia de Júpiter a la Tierra es de {round(dJT*conv)} millas, o {dJT} km.")
```
