# Ejercicio 2: Formateando Cadenas

```python
# Datos con los que vas a trabajar
name = "Luna"
gravity = 0.00162 # in kms
planet = "Tierra"

# Creamos el título
title = f"Datos sobre la gravedad en {name}\n"

# Creamos la plantilla 
facts = f"""-------------------------------------------------------------------------------
Nombre: {name}
Gravedad: {round(gravity*1000,2)}m/s2
Nombre del planeta: {planet}
"""
formato = title.title() + facts
print(formato)

name2 = "Ganímedes"
gravity2 = 0.00143 # in kms
planet2 = "Marte"

#Creamos una nueva plantilla
formato2 = formato.replace(name, name2)
formato2 = formato2.replace(str(round(gravity*1000,2)), str(round(gravity2*1000,2)))
formato2 = formato2.replace(planet, planet2)

print(formato2)
```
