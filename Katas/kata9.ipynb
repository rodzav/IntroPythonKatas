```python
# Función promedio 
def promedio(*nums):
    sumita=0
    for i in range(len(nums)):
        sumita = sumita + nums[i]
    return (sumita/len(nums))
```

```python
a=1
b=2
c=6
print(f"El promedio de los numeros {a}, {b} y {c} es de {promedio(a, b, c)}.""")
```

```python
# Actualiza la función
def prom_comb(tank1, tank2, tank3): #alaores en porcentaje
    return round(promedio(tank1,tank2,tank3),3)
```

```python
print(f"El promedo del combustible total en los tres tanques es de: {prom_comb(70,10,50)}%")
```

# Ejercicio 2: Trabajo con argumentos de palabra clave
## Ejercicio : Trabajar con argumentos de palabras clave en funciones

```python
# Función con un informe preciso de la misión. Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno

from datetime import timedelta, datetime

def arrival_time(hours2):
    arrival = datetime.now() + timedelta(hours=hours2)
    return arrival

def informe (dest, tiemp_dest, tank_e, tank_i):
    print(f"""INFORME DE COHETE
    DESTINO: {dest}
    TIEMPO HASTA EL DESTINO: {tiemp_dest} horas
    HORA DE PRELANZAMIENTO: {datetime.now()}
    FECHA DE LLEGADA: {arrival_time(tiemp_dest).strftime(f'%Y-%m-%d')}
    HORA DE LLEGADA: {arrival_time(tiemp_dest).strftime(f'%A %H:%M')}
    TANQUE EXTERNO: {tank_e}%
    TANQUE INTERNO: {tank_i}%""")
```

```python
informe("Ceres", 546, 82, 26)
```

```python
# Escribe tu nueva función de reporte considerando lo anterior
def informe_1 (dest, *horas, **reservas):
    print(f"""INFORME DE COHETE
    DESTINO: {dest}
    TIEMPO HASTA EL DESTINO: {sum(horas)} horas
    HORA DE PRELANZAMIENTO: {datetime.now()}
    FECHA DE LLEGADA: {arrival_time(sum(horas)).strftime(f'%Y-%m-%d')}
    HORA DE LLEGADA: {arrival_time(sum(horas)).strftime(f'%A %H:%M')}
    COMBUSTIBLE TOTAL: {round((sum(reservas.values()))/len(reservas),3)}%""")
```

```python
informe_1("Ganímedes", 20, 3, 152, interno=85, externo=100, externo2=100, externo3=50, emergencia=92)
```

```python
# Escribe tu nueva función
def informe_2 (dest, *horas, **reservas):
    print(f"""INFORME DE COHETE
    DESTINO: {dest}
    TIEMPO HASTA EL DESTINO: {sum(horas)} horas
    HORA DE PRELANZAMIENTO: {datetime.now()}
    FECHA DE LLEGADA: {arrival_time(sum(horas)).strftime(f'%Y-%m-%d')}
    HORA DE LLEGADA: {arrival_time(sum(horas)).strftime(f'%A %H:%M')}
    COMBUSTIBLE TOTAL: {round((sum(reservas.values()))/len(reservas),3)}%""")
    for i in reservas.keys():
        print(f"""        TANQUE {i.upper()}: {reservas[i]}""")
```

```python
informe_2("Ganímedes", 20, 3, 152, interno=85, externo=100, externo2=100, externo3=50, emergencia=92)
```
